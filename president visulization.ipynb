```python
import pandas as pd
from pandas import DataFrame, Series

import numpy as np

import matplotlib.pyplot as plt
import seaborn as sns
```

```python
presidents = pd.read_csv('presidents.csv')
```

```python
new_parties = presidents['party'].isin(['Democratic', 'Republican'])
```

```python
presidents_party = presidents[new_parties]
```

```python
sns.lmplot(data=presidents_party, x='year', y='gdp_bn', hue='party', fit_reg=False)
```

```python
sns.boxplot('gdp_bn', y='party', data=presidents_party)
```
